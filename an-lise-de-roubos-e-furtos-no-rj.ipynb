# **Análise de Roubos e Furtos de Celulares no Rio de Janeiro entre 01/2014 e 09/2024**

A criminalidade envolvendo roubos e furtos de celulares tem se tornado uma preocupação crescente em muitas cidades do Brasil, especialmente no Rio de Janeiro. Com o aumento desses delitos, não apenas as vítimas enfrentam prejuízos materiais, mas também um impacto emocional e psicológico considerável.

Estudos recentes, como o levantamento do Datafolha encomendado pelo Fórum Brasileiro de Segurança Pública (FBSP), revelam que 9,2% da população brasileira foi vítima de furto ou roubo de celular. Este dado preocupa, especialmente considerando que 53% dos entrevistados evitam determinados locais ou horários por medo de serem vítimas desses crimes. No Rio de Janeiro, a situação é particularmente crítica, com mais de 10.574 celulares roubados bloqueados apenas através do programa Celular Seguro, que visa proteger os consumidores e desincentivar a revenda ilegal de aparelhos roubados.

Neste contexto, o presente projeto visa analisar dados relacionados a roubos e furtos de celulares no Rio de Janeiro, utilizando informações de roubos por bairro para identificar padrões e tendências de criminalidade. A partir dessa análise, buscamos oferecer insights valiosos para entender melhor a dinâmica dessa problemática crescente, que afeta tanto a segurança pública quanto o comportamento da população.

Este repositório contém uma análise exploratória de dados sobre roubos e furtos de celulares no município do Rio de Janeiro. A análise por bairros será feita através das Unidades territoriais, com base no CISP (número da circunscrição onde ocorreu o fato). A fonte dos dados é o [Instituto de Segurança Pública do Rio de Janeiro](http://www.ispdados.rj.gov.br/estatistica.html).

**Objetivos da Análise:**

* Identificar a tendência de roubos e furtos ao longo dos anos.
* Determinar quais bairros possuem o maior número de ocorrências de roubos e furtos.
* Analisar os meses com maior incidência de roubos e furtos.

```python
# Importando os pacotes
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
```

```python
# Importando os dados
df_delegacias = pd.read_csv('/kaggle/input/roubos-e-furtos/delegacias.csv', sep=';', encoding='iso-8859-1')
df_furtos_roubos = pd.read_csv('/kaggle/input/roubos-e-furtos/roubos_furtos.csv', sep=';', encoding='iso-8859-1')
```

```python
# Padronizando as colunas para minúsculas
df_furtos_roubos.columns = df_furtos_roubos.columns.str.lower()
df_delegacias.columns = df_delegacias.columns.str.lower()
```

```python
# Filtrando os dados de furtos e roubos para o município do Rio de Janeiro
df_furtos_roubos_rio = df_furtos_roubos[df_furtos_roubos['munic'] == 'Rio de Janeiro']

# Realizando o merge entre os DataFrames
df_completo = pd.merge(df_furtos_roubos_rio, df_delegacias, on='cisp', how='left')

# Selecionando as colunas desejadas
df_final = df_completo[['mes', 'ano', 'roubo_celular', 'furto_celular', 'delegacia']]

# Alterando os nomes das colunas
df_final = df_final.rename(columns={
    'roubo_celular': 'roubos',
    'furto_celular': 'furtos',
})
```

```python
# Criando a coluna 'data' com formato 'YYYY-MM'
df_final['data'] = pd.to_datetime(df_final['ano'].astype(str) + '-' + df_final['mes'].astype(str), format='%Y-%m')

# Filtrando os dados entre janeiro de 2014 e setembro de 2024
start_date = '2014-01'
end_date = '2024-09'
df_final = df_final[(df_final['data'] >= start_date) & (df_final['data'] <= end_date)]
```

```python
# Verificando a quantidade de valores ausentes por coluna
df_final.isnull().sum()
```

```python
# Remover as linhas onde não há 'delegacia' 
df_final = df_final.dropna(subset=['delegacia'])
```

```python
# Estatísticas descritivas das colunas numéricas
df_final[['roubos', 'furtos']].describe()
```

```python
#Estatísticas descritivas de roubos por delegacia
df_grouped_roubos = df_final.drop(columns=['mes', 'ano', 'data', 'furtos']).groupby('delegacia').describe()
df_grouped_roubos
```

```python
#Estatísticas descritivas de furtos por delegacia
df_grouped_furtos = df_final.drop(columns=['mes', 'ano', 'data', 'roubos']).groupby('delegacia').describe()
df_grouped_furtos
```

# Tendência de roubos e furtos ao longo dos anos.

```python
# Agrupando os dados por mês e somando os valores de roubos e furtos de celular 
df_monthly = df_final.groupby('data')[['roubos', 'furtos']].sum().reset_index()

# Plotando a evolução mensal dos roubos e furtos
plt.figure(figsize=(12, 6))
plt.plot(df_monthly['data'], df_monthly['roubos'], label='Roubos de Celular', color='red', linestyle='-')
plt.plot(df_monthly['data'], df_monthly['furtos'], label='Furtos de Celular', color='blue', linestyle='-')
plt.title('Evolução Mensal de Roubos e Furtos de Celulares', fontsize=16)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Número de Ocorrências', fontsize=12)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.legend()
plt.tight_layout()
plt.show()
```

# Análise das delegacias com mais registros de roubos e furtos.

```python
# Média mensal de roubos por delegacia
df_roubos_bairros = df_final.groupby('delegacia')['roubos'].mean().reset_index()
df_roubos_bairros = df_roubos_bairros.sort_values(by='roubos', ascending=False)

print('Delegacias com mais ocorrências de roubos\n')
print(df_roubos_bairros.to_string(index=False))
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df_roubos_bairros, x='delegacia', y='roubos', palette='viridis')
plt.title('Média Mensal de Roubos por Delegacia', fontsize=16)
plt.xlabel('Delegacia', fontsize=12)
plt.ylabel('Média de Roubos', fontsize=12)
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Média mensal de furtos por delegacia
df_furtos_bairros = df_final.groupby('delegacia')['furtos'].mean().reset_index()
df_furtos_bairros = df_furtos_bairros.sort_values(by='furtos', ascending=False)


print('Delegacias com mais ocorrências de furtos\n')
print(df_furtos_bairros.to_string(index=False))
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df_furtos_bairros, x='delegacia', y='furtos', palette='viridis')
plt.title('Média Mensal de Furtos por Delegacia', fontsize=16)
plt.xlabel('Delegacia', fontsize=12)
plt.ylabel('Média de Furtos', fontsize=12)
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

A análise dos dados de roubos e furtos de celulares revela padrões interessantes de criminalidade na cidade do Rio de Janeiro. Ao observarmos as duas listas, vemos que as regiões com maior número de roubos e furtos não são as mesmas, embora algumas se sobreponham. Isso pode ser interpretado como uma diferença no perfil do crime em diferentes áreas da cidade.

**Bairros com mais ocorrências de Roubos**

A 34°DP - Bangu lidera com uma média de 53.01 roubos por mês, seguido pela 29°DP - Madureira com 45.58 roubos por mês.
A maior parte das regiões com alta taxa de roubos está localizada no subúrbio carioca, áreas com características de maior vulnerabilidade à violência, como a maior distância dos centros comerciais e turísticos e, muitas vezes, o deslocamento de pessoas em transporte público.

**Bairros com mais ocorrências de Furtos**

Já nos furtos, as áreas mais afetadas são as regiões mais turísticas da cidade, com destaque para Barra da Tijuca, Copacabana, Ipanema e Leblon, todas com grandes fluxos de pessoas e, portanto, mais suscetíveis a furtos devido ao descuido de vítimas e ao grande número de turistas.
Essas áreas também são conhecidas por grandes concentrações de eventos, como o carnaval, o que pode contribuir para o aumento da criminalidade por conta das multidões e da distração das pessoas.

# Análise dos meses com mais registros de roubos e furtos.

```python
# Médias de roubos e furtos de celulares por mês
df_monthly_mean = df_final.groupby('mes')[['roubos', 'furtos']].mean().reset_index()

# Mapeando os números dos meses para os nomes dos meses
df_monthly_mean['mes'] = df_monthly_mean['mes'].map({
    1: 'Janeiro', 2: 'Fevereiro', 3: 'Março', 4: 'Abril', 5: 'Maio', 6: 'Junho',
    7: 'Julho', 8: 'Agosto', 9: 'Setembro', 10: 'Outubro', 11: 'Novembro', 12: 'Dezembro'
})

print("Média de roubos de celulares por mês:")
print(df_monthly_mean[['mes', 'roubos']].to_string(index=False))

print("\nMédia de furtos de celulares por mês:")
print(df_monthly_mean[['mes', 'furtos']].to_string(index=False))
```

```python
# 3 meses com maiores médias de roubos e furtos

# Ordenando as médias de roubos e furtos em ordem decrescente
df_monthly_sorted_roubo = df_monthly_mean.sort_values(by='roubos', ascending=False)
df_monthly_sorted_furto = df_monthly_mean.sort_values(by='furtos', ascending=False)


print("3 meses com maiores médias de roubos de celulares:")
print(df_monthly_sorted_roubo[['mes', 'roubos']].head(3).to_string(index=False))

print("\n3 meses com maiores médias de furtos de celulares:")
print(df_monthly_sorted_furto[['mes', 'furtos']].head(3).to_string(index=False))
```

A média mensal de roubos de celulares mostra uma certa constância, com variações mínimas de um mês para outro. Apesar dessas flutuações, o número de roubos não mostra uma sazonalidade clara. Esse padrão sugere que os roubos de celulares acontecem de forma relativamente uniforme ao longo do ano, sem picos sazonais evidentes.

Em contraste, a média de furtos de celulares apresenta flutuações mais marcantes, com picos visíveis em certos meses do ano. Esses meses coincidem com grandes eventos da cidade, como o Carnaval (Fevereiro e Março) e o Rock in Rio (Setembro), eventos que atraem um grande número de turistas e locais para áreas movimentadas, o que pode explicar o aumento na incidência de furtos.

# Conclusão

A análise revela a importância de abordagens distintas no combate aos furtos e roubos de celulares. No caso dos furtos, é crucial focar em estratégias específicas durante grandes eventos e períodos de grande concentração de pessoas. Aumento da segurança, aliado a campanhas educativas que alertem sobre os riscos de distração e ofereçam orientações de como se proteger, pode fazer a diferença. Já para os roubos, é necessário um esforço contínuo e de longo prazo, com investimentos sólidos em segurança pública, como o reforço do policiamento e a fiscalização mais rigorosa nas áreas mais vulneráveis.
